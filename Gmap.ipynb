```python
class FatalityExplorer(object):
    """
    Jupyter widget for exploring the Austin Trafic Fatality dataset.
    
    The user uses the slider to choose a year. This renders a heatmap
    of fatality locations in that year.
    """
    def __init__(self, fatality_locations):
        self._df = fatality_locations
        self._heatmap = None
        self._slider = None
        initial_year = min(self._df['year'])
        title_widget = widgets.HTML(
            '<h3>Trafic Fatalities in City Austin, by year</h3>'
            '<h4>Data from Austin Police Department</h4>'
        )
        map_figure = self._render_map(initial_year)
        controls = self._render_controls(initial_year)
        self._container = widgets.VBox([title_widget, controls, map_figure])
        

    def render(self):
        display(self._container)

    def _on_year_change(self, change):
        year = self._slider.value
        self._heatmap.locations = self._locations_for_year(year)
        self._total_box.value = self._total_fatalities_text_for_year(year)
        return self._container

    def _render_map(self, initial_year):
        fig = gmaps.figure()
        self._heatmap = gmaps.heatmap_layer(
            self._locations_for_year(initial_year),
#             max_intensity=100,
#              point_radius=5
        )
        fig.add_layer(self._heatmap)
        return fig

    def _render_controls(self, initial_year):
        self._slider = widgets.IntSlider(
            value=initial_year,
            min=min(self._df['year']),
            max=max(self._df['year']),
            description='year',
            continuous_update=False
        )

        self._total_box = widgets.Label(
            value=self._total_fatalities_text_for_year(initial_year)
        )
        self._slider.observe(self._on_year_change, names='value')
        controls = widgets.HBox(
            [self._slider, self._total_box],
            layout={'justify_content': 'space-between'}
        )
        return controls
    
    def _locations_for_year(self, year):
        return self._df[self._df['year'] == year][["Y COORD", "X COORD"]]

    def _total_fatalities_for_year(self, year):
        return int(self._df[self._df['year'] == year]['year'].count())

    def _total_fatalities_text_for_year(self, year):
        return '{} Traffice Fatalities'.format(self._total_fatalities_for_year(year))

FatalityExplorer(fatality_locations).render()
```

```python
class TypesExplorer(object):

    def __init__(self, fatality18_types):

        """
        Jupyter widget for exploring KFC and Starbucks outlets
        Using checkboxes, the user chooses whether to include
        Starbucks, KFC outlets, both or neither.
        """
        self._df = fatality18_types
        self._symbol_layer = None
        self._motorvehicle_symbols = self._create_symbols_for_chain(
            'motorvehicle', 'red')
        self._motorcycle_symbols = self._create_symbols_for_chain(
            'motorcycle', 'blue')
        self._pedestrian_symbols = self._create_symbols_for_chain(
            'pedestrian', 'black')
        self._bicycle_symbols = self._create_symbols_for_chain(
            'bicycle', 'yellow')
        title_widget = widgets.HTML(
            '<h3>Explore Vehicle Types of Traffic Fatalities in City Austin 2018</h3>'
            '<h4>Data from Austin Police Department</h4>'
        )

        controls = self._render_controls(True, True, True, True)
        map_figure = self._render_map(True, True, True, True)
        self._container = widgets.VBox(
            [title_widget, controls, map_figure])

    def render(self):
            """ Render the widget """
            display(self._container)

    def _render_map(
            self, 
            initial_include_motorvehicle, 
            initial_include_motorcycle,
            initial_include_pedestrian, 
            initial_include_bicycle
        ):
            """ Render the initial map """
            fig = gmaps.figure(layout={'height': '500px'})
            symbols = self._generate_symbols(True, True, True, True)
            self._symbol_layer = gmaps.Markers(markers=symbols)
            fig.add_layer(self._symbol_layer)
            return fig

    def _render_controls(
            self,
            initial_include_motorvehicle, 
            initial_include_motorcycle,
            initial_include_pedestrian, 
            initial_include_bicycle
        ):
            """ Render the checkboxes """
            self._motorvehicle_checkbox = widgets.Checkbox(
                value=initial_include_motorvehicle,
                description='motorvehicle'
            )
            self._motorcycle_checkbox = widgets.Checkbox(
                value=initial_include_motorcycle,
                description='motorcycle'
            )
            self._pedestrian_checkbox = widgets.Checkbox(
                value=initial_include_pedestrian,
                description='pedestrian'
            )
            self._bicycle_checkbox = widgets.Checkbox(
                value=initial_include_bicycle,
                description='bicycle'
            )

            self._motorvehicle_checkbox.observe(
                self._on_controls_change, names='value')
            self._motorcycle_checkbox.observe(
                self._on_controls_change, names='value')
            self._pedestrian_checkbox.observe(
                self._on_controls_change, names='value')
            self._bicycle_checkbox.observe(
                self._on_controls_change, names='value')
            
           
            controls = widgets.VBox(
                [self._motorvehicle_checkbox, 
                 self._motorcycle_checkbox,
                 self._pedestrian_checkbox,
                 self._bicycle_checkbox,
                ])
            return controls

    def _on_controls_change(self, obj):
            """
            Called when the checkboxes change
            This method builds the list of symbols to include on the map,
            based on the current checkbox values. It then updates the
            symbol layer with the new symbol list.
            """

            include_motorvehicle = self._motorvehicle_checkbox.value
            include_motorcycle = self._motorcycle_checkbox.value
            include_pedestrian = self._pedestrian_checkbox.value
            include_bicycle = self._bicycle_checkbox.value
            
            symbols = self._generate_symbols(
                include_motorvehicle, 
                include_motorcycle,
                include_pedestrian,
                include_bicycle)
            # Update the layer with the new symbols:
            self._symbol_layer.markers = symbols

    def _generate_symbols(
                self, 
                include_motorvehicle, 
                include_motorcycle,
                include_pedestrian,
                include_bicycle
            ):
                """ Generate the list of symbols to includs """
                symbols = []
                if include_motorvehicle:
                    symbols.extend(self._motorvehicle_symbols)
                if include_motorcycle:
                    symbols.extend(self._motorcycle_symbols)
                if include_pedestrian:
                    symbols.extend(self._pedestrian_symbols)
                if include_bicycle:
                    symbols.extend(self._bicycle_symbols)
                return symbols

    def _create_symbols_for_chain(self, chain, color):
                chain_df = self._df[self._df['Type'] == chain]
                symbols = [
                    gmaps.Symbol(
                        location=(latitude, longitude),
                        stroke_color = color,
                        fill_color = color,
                        scale=2
                    )
                    for latitude, longitude in
                    zip(chain_df["Y COORD"], chain_df["X COORD"])
                ]
                return symbols           

TypesExplorer(fatality18_types).render()
```
